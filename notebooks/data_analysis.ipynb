```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Load the CSV file properly
file_path = 'parsed_reward_data.csv'
data = pd.read_csv(file_path)
```

```python
# Extract episodes and rewards
episodes = data['Episode']
rewards = data['Reward']
```

```python
# Create windowed moving average
window_size = 50  # you can adjust
moving_avg = rewards.rolling(window=window_size, min_periods=1).mean()
```

```python
# Plot
plt.figure(figsize=(12, 6))
plt.plot(episodes, rewards, label='Raw Reward', alpha=0.3)
plt.plot(episodes, moving_avg, label=f'{window_size}-Episode Moving Average', linewidth=2)
plt.xlabel('Episode')
plt.ylabel('Reward')
plt.title('Reward Progression Over Time')
plt.legend()
plt.grid(True)
plt.show()
```
